1.) Use matplotlib to plot the following equation:

- y= x^2âˆ’x+2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
x = list(range(-25,25))

def y_plot(x):
    y = [(x**2)- x + 2 for x in x]
    return y

y = y_plot(x)

plt.plot(x, y)

# arrowprops == properties of the arrow
# xy == arrow tip coords
# xytext == location of body of arrow

# Find more @ https://matplotlib.org/2.0.2/users/annotations.html

plt.annotate('Point Origin (0,0)', xy=(0,0), xytext=(0, 100),
arrowprops=dict(facecolor='black', shrink=1))


plt.show()
```

- y = sqrt(x)
- y = x^3

You can use functions from the math module to help implement some of the equations above.

3.) Combine the figures you created in the last step into one large figure with 4 subplots.

4.) Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = list(range(100))

y1= [math.sqrt(x) for x in x]
y2= [(x*x*x) for x in x]
y3= [(2**x) for x in x]
y4= [(1/(x+1)) for x in x]



plt.figure(figsize = (8,6))
sets = [y1, y2, y3, y4]

for i in range(len(sets)):
    plt.subplot(2,2,i+1)
    plt.plot(sets[i])
    


    
```

```python
plt.figure(figsize = (8,8))

x = list(range(0,4))

y1= [math.sqrt(x) for x in x]
y2= [(x*x*x) for x in x]
y3= [(2**x) for x in x]
y4= [(1/(x+1)) for x in x]

plt.title('Combined Figures', fontsize = 22)

plt.plot(x, y1, c = 'orange', label = 'Square root of x')
plt.plot(x, y2, c = 'red', label = 'x^3')
plt.plot(x, y3, c = 'blue', label = '2^x')
plt.plot(x, y4, c = 'purple', label = '1/(x+1)')
         
plt.legend()
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```
