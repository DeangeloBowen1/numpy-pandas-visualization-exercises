How many unique titles are in the titles DataFrame?

```python
query = '''
SELECT COUNT(DISTINCT title) as 'uniqe_titles'
FROM titles 
'''

titles = pd.read_sql(query, employees)
titles
```

What is the oldest date in the to_date column?

```python
query = '''
SELECT MIN(to_date) as oldest_date
FROM titles 
'''

max_date = pd.read_sql(query, employees)
max_date
```

What is the most recent date in the to_date column?

```python
query = '''
SELECT MAX(from_date) as recent_date,
to_date
FROM titles 
GROUP BY to_date
LIMIT 1
'''

recent_date = pd.read_sql(query, employees)
recent_date
```

```python
import pandas as pd
import numpy as np

# Create the users DataFrame.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

What is the result of using a right join on the DataFrames?

```python
# Perform an right join specifying the left and right DataFrame keys.

users.merge(roles, left_on='role_id',
            right_on='id', how='right', indicator=True)
```

What is the result of using an outer join on the DataFrames?

```python
# Perform an outer join specifying the left and right DataFrame keys.

users.merge(roles ,left_on='role_id',
            right_on='id', how='outer', indicator=True)
```

- 7.) Write the code necessary to create a cross tabulation of the number of titles by department. (Hint: this will involve a combination of SQL code to pull the necessary data and python/pandas code to perform the manipulations.)

```python
dept_df = pd.read_sql('''
SELECT *
FROM employees
JOIN titles USING (emp_no)
JOIN dept_emp USING (emp_no)
JOIN departments d USING (dept_no)

WHERE titles.to_date = "9999-01-01" 
AND dept_emp.to_date = "9999-01-01"
''', employees)
dept_df
```

```python
pd.crosstab(dept_df.title, dept_df.dept_name)
```
